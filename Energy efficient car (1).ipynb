```python
y=pd.read_csv(r'C:\\Quarter 1\\6020 Predictive analysis\\Module 2 Linear regression\\\Final\\car.csv')
```

```python
y = y.loc[~(y['Horsepower'].isin(['?']))]
y['Horsepower'] = pd.to_numeric(y['Horsepower'], downcast='integer', errors='coerce')
y
```

```python
from sklearn.model_selection import train_test_split



#Split
x = y.iloc[:, range(1,8)]
z = y[['MPG']]

x_train, x_test, z_train, z_test = train_test_split(x, z, test_size=0.20)

import statsmodels.api as sm

lr = sm.add_constant(x_train) 
lr_model = sm.OLS(z_train, lr) 
lr_fit = lr_model.fit()
print(lr_fit.summary())
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Feature extraction
test = SelectKBest(score_func=chi2, k=5)
x=y[['US Made','Model Year','Weight','Displacement','Horsepower','Cylinders','Acceleration']]
z=z.astype('int')
fit = test.fit(x, z)

# Summarize scores
np.set_printoptions(precision=1)
print(fit.scores_)

features = fit.transform(x)
# Summarize selected features
print(features[0:5,:])
#get_feature_names_out(input_features=None)
#print(features.get_feature_names())
mask = test.get_support() #list of booleans
new_features = [] # The list of your K best features

for bool, feature in zip(mask,['US Made','Model Year','Weight','Displacement','Horsepower','Cylinders','Acceleration']):
    if bool:
        new_features.append(feature)
        
print(new_features)
```
